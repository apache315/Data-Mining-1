```python

def count_outliers_std(df, columns, threshold=3):
    outliers_counts = {}
    for col in columns:
        std = df[col].std()
        mean = df[col].mean()
        lower_bound = mean - threshold * std
        upper_bound = mean + threshold * std
        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()
        outliers_counts[col] = outliers_count
    return outliers_counts


# Seleziona le colonne numeriche continue
numeric_columns = ['duration_ms', 'popularity', 'danceability', 'energy', 'loudness', 
                   'speechiness', 'acousticness', 'instrumentalness', 'liveness', 
                   'valence', 'tempo', 'features_duration_ms', 'n_beats', 'n_bars', 
                   'popularity_confidence']

# Conta gli outlier utilizzando il metodo della deviazione standard
outliers_counts = count_outliers_std(df, numeric_columns)

# Stampa il conteggio degli outlier per ogni colonna
print("Il conteggio degli outlier per ogni colonna :")
for col, count in outliers_counts.items():
    print(f"{col}: {count}")
```

```python

def count_outliers_iqr(df, columns):
    outliers_counts = {}
    for col in columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()
        outliers_counts[col] = outliers_count
    return outliers_counts


# Seleziona le colonne numeriche continue
numeric_columns = ['duration_ms', 'popularity', 'danceability', 'energy', 'loudness', 
                   'speechiness', 'acousticness', 'instrumentalness', 'liveness', 
                   'valence', 'tempo', 'features_duration_ms', 'n_beats', 'n_bars', 
                   'popularity_confidence']

# Conta gli outlier utilizzando il metodo dei quartili (IQR)
outliers_counts = count_outliers_iqr(df, numeric_columns)

# Stampa il conteggio degli outlier per ogni colonna
print("Il conteggio degli outlier per ogni colonna :")
for col, count in outliers_counts.items():
    print(f"{col}: {count}")
```
